```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Input.csv",encoding='latin-1')
```

```python
# Checking missing values

df.isnull().sum()
```

### Handiling Missing value

```python

df['Email'] = df['Email'].fillna('unknown@gmail.com') #missing value correction
df['Age'] = df['Age'].fillna(df['Age'].mean()) # Missing Value Correction
df['City'] = df['City'].fillna('unknown') #missing value correction

df
```

```python
# Fixing messy email
import re
def is_valid_email(email):
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return bool(re.match(pattern, str(email)))

# Store invalid rows
invalid_rows = df[~df['Email'].apply(is_valid_email)]

# Save invalid rows (optional but good)
invalid_rows.to_csv("invalid_data.csv", index=False)

# Keep only valid rows
df = df[df['Email'].apply(is_valid_email)]
```

```python
# Fixing age with inproper number
for idx, i in enumerate(df['Age']):
    if i < 0 or i > 120:
        df.at[idx, 'Age'] = df['Age'].mean()
df
```

```python
df['Email'] = df['Email'].str.lower() # Validate Email
df
```

```python
# Standardize text fields
df['Name'] = df['Name'].str.title()
df['Email'] = df['Email'].str.lower()
```

```python
# Derived fields

def age_group(age):
  if age <= 18:
    return 'Teen'
  elif age <= 30:
    return 'Young'
  elif age <= 50:
    return 'Middle'
  else:
    return 'Old'
df['Age Group'] = df['Age'].apply(age_group)
df
```

```python
# export in output.csv
df.to_csv('output.csv',index = False)
```

```python
chunks = pd.read_csv("/content/Input.csv", chunksize=1000)
cleaned_chunks = []
for chunk in chunks:
    # Apply your cleaning steps here

    chunk['Email'] = chunk['Email'].str.lower()

    chunk = chunk[chunk['Email'].apply(is_valid_email)]

    chunk.loc[(chunk['Age'] < 0) | (chunk['Age'] > 120), 'Age'] = chunk['Age'].mean()

    cleaned_chunks.append(chunk)

# Combine all chunks
df = pd.concat(cleaned_chunks)

# Save final cleaned data
df.to_csv("cleaned_data.csv", index=False)
```
